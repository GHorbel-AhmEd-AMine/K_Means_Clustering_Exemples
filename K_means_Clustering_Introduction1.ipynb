# Introduction au K_Means_Clustering

"""

Clustering is the process of dividing the entire data into groups (also known as clusters) based on the patterns in the data.

"""

Le regroupement K-means est un type d'apprentissage non supervisé, qui est utilisé lorsque vous avez des données non étiquetées (c'est-à-dire des données sans catégories ou groupes définis). Le but de cet algorithme est de trouver des groupes dans les données, le nombre de groupes étant représenté par la variable K. L'algorithme fonctionne de manière itérative pour attribuer chaque point de données à l'un des K groupes en fonction des caractéristiques fournies. Les points de données sont regroupés en fonction de la similarité des caractéristiques.Les résultats de l'algorithme de regroupement des K moyennes sont les suivants :

Les centroïdes des K groupes, qui peuvent être utilisés pour étiqueter les nouvelles données

Étiquettes pour les données de formation (chaque point de données est attribué à un seul groupe)

Plutôt que de définir des groupes avant d'examiner les données, le regroupement vous permet de trouver et d'analyser les groupes qui se sont formés de manière organique.


La section "Choisir K" ci-dessous décrit comment le nombre de groupes peut être déterminé:  

Chaque centroïde d'une cluster est un ensemble de valeurs caractéristiques qui définissent les groupes résultants. L'examen des poids des caractéristiques des centroïdes peut être utilisé pour interpréter qualitativement le type de groupe que chaque groupe représente.  


## Utilisations commerciales: 


L'algorithme de regroupement de K-means est utilisé pour trouver des groupes qui n'ont pas été explicitement étiquetés dans les données. Cela peut être utilisé pour confirmer les hypothèses des entreprises sur les types de groupes existants ou pour identifier des groupes inconnus dans des ensembles de données complexes. Une fois l'algorithme exécuté et les groupes définis, toute nouvelle donnée peut être facilement attribuée au groupe correct.

Il s'agit d'un algorithme polyvalent qui peut être utilisé pour tout type de regroupement. Voici quelques exemples de cas d'utilisation :

Segmentation comportementale :

Segmentation par historique d'achat
Segment par activités sur l'application, le site web ou la plate-forme
Définir les personnalités en fonction des intérêts
Créer des profils basés sur le suivi des activités

Catégorisation de l'inventaire :

Inventaire de groupe par activité de vente
Groupement des stocks par métrique manufacturière

Tri des mesures des capteurs :

Détecter les types d'activité dans les capteurs de mouvement
Images de groupe
Audio séparé
Identifier les groupes dans la surveillance de la santé

Détection de robots ou d'anomalies :

Séparer les groupes d'activités valides des robots
Regrouper les activités valides afin d'améliorer la détection des valeurs aberrantes
En outre, le suivi des changements de groupes dans le temps peut être utilisé pour détecter des changements significatifs dans les données.
